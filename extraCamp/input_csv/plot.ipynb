```python
# Read flood_levels.csv and plot flood levels over time in a heatmap
import pandas as pd
import matplotlib.pyplot as plt
# Load the data
df = pd.read_csv('flood_level.csv')

# columns = ['#day', 'location1', 'location2', 'location3', 'location4', etc.]
# Set the index to the day column
df.set_index('#day', inplace=True)
# Plot the heatmap
plt.figure(figsize=(10, 6))
plt.imshow(df.T, aspect='auto', cmap='viridis', interpolation='nearest')
plt.colorbar(label='Flood Level')
plt.xticks(range(len(df.index)), df.index, rotation=45)
plt.yticks(range(len(df.columns)), df.columns)
plt.title('Flood Levels Over Time')
plt.xlabel('Day')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```
